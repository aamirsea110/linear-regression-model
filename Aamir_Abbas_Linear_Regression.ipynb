```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import LabelEncoder
```

```python
df_train = pd.read_csv('train_dataset.csv', index_col=0)
df_train.sort_index(inplace=True)
df_train.drop(columns=['communications', 'operating system', 'input devices', 'multimedia', 'state'], inplace=True)
# in this line, first select the most frequent value, replace nan values with it, then show value counts
df_train['RAM size'] = df_train['RAM size'].fillna(df_train['RAM size'].value_counts().index[0]).apply(lambda x: int(x[:-2]))
df_train.dropna(inplace=True)
df_train['CPU cores'] = df_train['CPU cores'].astype(np.int64)
df_train['resolution (px)'] = df_train['resolution (px)'].apply(lambda x: int(x.split(' x ')[0]) * int(x.split(' x ')[1]))
df_train
```

```python
df_test = pd.read_csv('test_dataset.csv', index_col=0)
df_test.sort_index(inplace=True)
df_test.drop(columns=['communications', 'operating system', 'input devices', 'multimedia', 'state'], inplace=True)
# in this line, first select the most frequent value, replace nan values with it, then show value counts
df_test['RAM size'] = df_test['RAM size'].fillna(df_test['RAM size'].value_counts().index[0]).apply(lambda x: int(x[:-2]))
df_test.dropna(inplace=True)
df_test['CPU cores'] = df_test['CPU cores'].astype(np.int64)
df_test['resolution (px)'] = df_test['resolution (px)'].apply(lambda x: int(x.split(' x ')[0]) * int(x.split(' x ')[1]))
df_test
```

```python
df_train = pd.get_dummies(df_train)
df_train
```

```python
df_test = pd.get_dummies(df_test)
columns_not_in_test = list(set(df_train.columns).difference(set(df_test.columns)))
df_test[columns_not_in_test] = 0
df_test
```

```python
X_train = df_train.drop(columns='buynow_price').values
y_train = df_train['buynow_price'].values
X_test = df_test.drop(columns='buynow_price').values
y_test = df_test['buynow_price'].values
print('X_train:', X_train.shape)
print('y_train:', y_train.shape)
print('X_test:', X_test.shape)
print('y_test:', y_test.shape)
```

```python
reg = LinearRegression()
reg.fit(X_train, y_train)
print(reg.score(X_test, y_test))
```
